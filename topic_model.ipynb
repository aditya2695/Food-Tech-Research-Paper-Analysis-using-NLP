```python
import pandas as pd
import gensim
from gensim import corpora

```

```python
journal_df = pd.read_csv('data/journals.csv')
journal_df.head()
```

```python
# Drop rows where the "abstract" column contains NaN values
journal_df = journal_df.dropna(subset=['abstract'])
journal_df[journal_df['abstract'].isna()] 
```

```python
def get_topics(df):

    # Preprocess the text data
    processed_data = [document.lower().split() for document in df['abstract']]

    # Create a dictionary from the preprocessed data
    dictionary = corpora.Dictionary(processed_data)

    # Create a document-term matrix
    corpus = [dictionary.doc2bow(document) for document in processed_data]

    # Perform topic modeling using LDA
    num_topics = 5  # Adjust the number of topics as needed
    lda_model = gensim.models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)

    # Print the topics and their top keywords
    for topic_id, topic in lda_model.show_topics(num_topics=num_topics):
        print(f"Topic #{topic_id + 1}:")
        print(topic)
        print()

    # Assign topics to the documents
    topic_assignments = [lda_model.get_document_topics(doc) for doc in corpus]

    return topic_assignments
```

```python
topic_assignments = get_topics(journal_df)

# Add topic assignments to the DataFrame
journal_df['topic_assignments'] = topic_assignments

journal_df.head(3)
```
